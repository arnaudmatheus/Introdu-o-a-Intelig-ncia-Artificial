#### Foi percebido que os valores column 1, column 2, column 3, column 4 possuem valores em branco portanto precisam ser removidos, bem como num_reactions por ser altamente correlacionado com num_likes

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

for columns in data.columns:
     if data[columns].dtypes == 'object':
        data[columns] = le.fit_transform(data[columns])
data=data.drop(['Column1','Column2','Column3','Column4','num_reactions'],axis=1)
data
```

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
rs = MinMaxScaler()
pca = PCA(n_components=2).fit(data)
pca_2d = pca.transform(data)
final = pd.DataFrame()



for k in range(1,20):
    kmeans = KMeans(n_clusters=k,random_state = True)
    kmeans.fit(data)
    labels = kmeans.predict(data)
    centroids = kmeans.cluster_centers_
    final['Clusters',k-1] = labels
    
```

#### Implementação do k-means para valores diferentes de clusters para avaliar qual número de cluster possui o melhor agrupamento, nesse caso o melhor número de clusters é 4.

```python
for i in range(0,19):
    print("\n",final[('Clusters',i)].value_counts())
    
```
